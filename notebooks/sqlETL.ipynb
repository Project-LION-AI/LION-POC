```python
import pandas as pd
import numpy as np

def bubble_sort(list):
    for i in range(len(list)):
        for j in range(len(list)-1):
            if list[j] > list[j+1]:
                list[j], list[j+1] = list[j+1], list[j]
    return list

list = np.random.randint(0,100,10)
print(list)
print(bubble_sort(list))
```
